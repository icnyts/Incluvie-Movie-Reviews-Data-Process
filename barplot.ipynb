```python
#import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(9,9))
N = 5
all_review =(2342,3425,2654,3212,1988)
diversity = (124,245,212,321,85)
no_diversity = (2342-124,3425-245,2654-212,3212-321.05,1988-85)
percentage = []
for i,j in zip(diversity,no_diversity):
    percentage.append('%.2f' % float(i/j))
ind = range(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, diversity)
for a,b in zip(ind, diversity):
    plt.text(a, b, str(b), horizontalalignment='center',
          verticalalignment='bottom',size=15)
p2 = plt.bar(ind, no_diversity,
             bottom=diversity)

for a,b in zip(ind, all_review):
    plt.text(a, b, str(b),horizontalalignment='center',
          verticalalignment='bottom',size=15)
    
plt.ylabel('numbers of reviews')
plt.title('distribution of diversities for movies')
plt.xticks(ind, ('MV1', 'MV2', 'MV3', 'MV4', 'MV5'))
plt.yticks(range(0,3600,200))
plt.legend((p1[0], p2[0]), ('diversity', 'no diversity'),fontsize = 12.5,loc="upper right")

plt.show()
```
