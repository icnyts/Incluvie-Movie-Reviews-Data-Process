```python
# read csv file to a list of dictionaries
import csv
input_file = csv.DictReader(open("keywords.csv"))
l = list(input_file)
l
```

```python
# change the invalid character in the strings
for dic in l:
    dic['Name'] = dic.pop('\ufeffName')
    for key, value in dic.items():
        dic[key] = value.replace('\xa0','')
```
