```python
import matplotlib.pyplot as plt
 
# Data to plot
labels = ['diversity', 'non_diversity']
sizes = [4,96]
colors = ['lightcoral', 'lightskyblue']
explode = (0.3, 0)  # explode 1st slice
patches, text1, text2 = plt.pie(sizes, autopct='%1.1f%%',colors=colors,explode=explode,shadow=True, startangle=90)
plt.legend(patches, labels, loc="best")
plt.axis('equal')
plt.tight_layout()
plt.title("metioned diversity percentage")
plt.show()
```
