```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
race_keyword = [["american indian", "alaska native"], 
                ["asian"], 
                ["black", "african american"],
                ["native hawaiian", "other pacific islander"],
                ["white"]]
```

```python
import csv
all_reviews = list();
with open('starwars8.csv', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    i = 0
    for row in csv_reader:
        all_reviews.append(dict(row))
        #i += 1
        if (i >= 100):
            break
        
```

```python
# we want to keep a list of dict
# for each dict, it should follow the form of {"id": , "title_rds":, "title_ads" , "text_rds": , "text_ads": }
ds_reviews = list()

review_mentioned = 0
total_mentioned_text = 0
total_mentioned_title = 0
total_words = 0
total_title = 0

for review in all_reviews:
    flag = False
    text = review.get("review")
    title = review.get("title")
    
    ds_review = {"id":review.get("id")}
    
    exclude = set(string.punctuation)
    clear_text = ''.join(ch.lower() for ch in text if ch not in exclude)
    clear_title = ''.join(ch.lower() for ch in title if ch not in exclude)
    clear_text_list = clear_text.split(" ")
    clear_title_list = clear_title.split(" ")
    #print(clear_text)
    
    text_ds = 0
    for word in clear_text_list:
        for each in race_keyword:
            if word in each:
                flag = True
                print("catch word")
                text_ds += 1
    total_mentioned_text += text_ds
    total_words += len(clear_text_list)
    ds_review["text_ads"] = text_ds
    ds_review["text_rds"] = text_ds/len(clear_text_list)
    
    title_ds = 0
    for word in clear_title_list:
        for each in race_keyword:
            if word in each:
                flag = True
                print("catch title")
                title_ds += 1
                
    total_title += len(clear_title_list)
    total_mentioned_title += title_ds
    ds_review["title_ads"] = title_ds
    ds_review["title_rds"] = title_ds/len(clear_title_list)
    
    if flag:
        review_mentioned += 1
    ds_reviews.append(ds_review)
    
```
